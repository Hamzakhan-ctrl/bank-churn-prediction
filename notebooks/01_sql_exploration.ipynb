```python
import sqlite3
import pandas as pd

df = pd.read_csv('../data/Churn_Modelling.csv')
conn = sqlite3.connect(':memory:')
df.to_sql('churn', conn, index=False)

# Q1 - Overall churn rate
print("=== CHURN RATE ===")
print(pd.read_sql("SELECT ROUND(AVG(Exited)*100, 2) AS Churn_Rate_Percent FROM churn", conn))

# Q2 - Churn by state
print("\n=== CHURN BY STATE ===")
print(pd.read_sql("SELECT Geography, COUNT(*) AS Total, SUM(Exited) AS Churned, ROUND(AVG(Exited)*100,2) AS Churn_Pct FROM churn GROUP BY Geography ORDER BY Churn_Pct DESC", conn))

# Q3 - Churn by gender
print("\n=== CHURN BY GENDER ===")
print(pd.read_sql("SELECT Gender, ROUND(AVG(Exited)*100,2) AS Churn_Pct FROM churn GROUP BY Gender", conn))

# Q4 - Churn by number of products
print("\n=== CHURN BY PRODUCTS ===")
print(pd.read_sql("SELECT NumOfProducts, ROUND(AVG(Exited)*100,2) AS Churn_Pct FROM churn GROUP BY NumOfProducts", conn))

# Q5 - Avg balance churned vs retained
print("\n=== AVG BALANCE ===")
print(pd.read_sql("SELECT Exited, ROUND(AVG(Balance),2) AS Avg_Balance FROM churn GROUP BY Exited", conn))
```
