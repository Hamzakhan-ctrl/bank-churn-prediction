```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('../data/Churn_Modelling.csv')
sns.set_style("darkgrid")

# Plot 1 - Churn Distribution
plt.figure(figsize=(6,4))
df['Exited'].value_counts().plot(kind='bar', color=['steelblue','tomato'])
plt.title('Churn Distribution')
plt.xticks([0,1], ['Retained','Churned'], rotation=0)
plt.ylabel('Count')
plt.tight_layout()
plt.show()

# Plot 2 - Churn by State
plt.figure(figsize=(8,4))
state_churn = df.groupby('Geography')['Exited'].mean()*100
state_churn.sort_values().plot(kind='barh', color='steelblue')
plt.title('Churn Rate by State (%)')
plt.xlabel('Churn %')
plt.tight_layout()
plt.show()

# Plot 3 - Age Distribution
plt.figure(figsize=(8,4))
df[df['Exited']==1]['Age'].hist(alpha=0.7, color='tomato', label='Churned', bins=30)
df[df['Exited']==0]['Age'].hist(alpha=0.7, color='steelblue', label='Retained', bins=30)
plt.title('Age Distribution by Churn')
plt.xlabel('Age')
plt.legend()
plt.tight_layout()
plt.show()

# Plot 4 - Churn by Gender
plt.figure(figsize=(6,4))
df.groupby('Gender')['Exited'].mean().mul(100).plot(kind='bar', color=['steelblue','tomato'])
plt.title('Churn Rate by Gender (%)')
plt.xticks(rotation=0)
plt.ylabel('Churn %')
plt.tight_layout()
plt.show()

# Plot 5 - Correlation Heatmap
plt.figure(figsize=(10,6))
sns.heatmap(df.select_dtypes(include='number').corr(), annot=True, fmt='.2f', cmap='Blues')
plt.title('Correlation Heatmap')
plt.tight_layout()
plt.show()

print("✅ EDA Complete!")
```
